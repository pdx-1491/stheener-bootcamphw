```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "../Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)

```

## Player Count

* Display the total number of players


```python
# checking to see if the dataframe loaded
# purchase_data.head()

# checking to see if there are any missing rows
# purchase_data.count()

# checking the datatypes of the columns
# purchase_data.dtypes

# using the SN column to get and display the total player count in a series
player_count = purchase_data['SN'].nunique()

# displaying the unique number of players in a dataframe
player_count = pd.DataFrame({'Total Players': [player_count]})
player_count

```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# create a collection of series that answer each calc. question and them combine them in a dataframe

# creating a series for 'Number of Unqiue Items'
unique_items = purchase_data['Item Name'].nunique()
# unique_items = pd.DataFrame('Item Name').nunique()

# creating a series for 'Average Price'
average_price = purchase_data['Price'].mean()
# average_price

# creating a series for 'Number of Purchases'
number_of_purchases = purchase_data['Purchase ID'].nunique()
# number_of_purchases

# creating a series for 'Total Revenue'
total_revenue = purchase_data['Price'].sum()
# total_revenue

# combining all the series into a new dataframe called summary_df
summary_df = pd.DataFrame({'Number of Unique Items': [unique_items], 'Average Price': [average_price],
                           'Number of Purchases': [number_of_purchases], 'Total Revenue': [total_revenue]})
# mapping 'Average Price' and 'Total Revenue' so they display as currency and rounded to the 2nd place
summary_df['Average Price'] = summary_df['Average Price'].map("${:.2f}".format)
summary_df['Total Revenue'] = summary_df['Total Revenue'].map("${:.2f}".format)
# final dataframe
summary_df

# help resolving creating dataframe from series cited here: https://stackoverflow.com/questions/17839973/constructing-pandas-dataframe-from-values-in-variables-gives-valueerror-if-usi

```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed


```python
# making the player_demographics dataframe using loc from the orginal purchase_data dataframe
player_demographics = purchase_data.loc[:, ["Gender", "SN", "Age"]]

# dropping duplicates since there are multiple entires from the same players in the original dataframe
player_demographics = player_demographics.drop_duplicates()

# getting a count of the unique values in the Gender column from the player_demogrpahics dataframe
player_demographics_count = player_demographics['Gender'].value_counts()

# making total_players from the player_count dataframe (used earlier) and iloc to determine total number of players
total_players = player_count.iloc[0]['Total Players']

# getting the percentage of genders from the demographics_count and total_players dataframe
player_demographics_percent = player_demographics_count/total_players * 100

# making a new dataframe to combine the dataframes
# use mapping to properly display the 'Percentage of Players' column
player_demographics = pd.DataFrame({'Total Count': player_demographics_count, 'Percentage of Players': player_demographics_percent})
player_demographics['Percentage of Players'] = player_demographics['Percentage of Players'].map("{:.2f}%".format)
player_demographics

```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# series called purchase_count with 'Gender', 'Price' columns from purchase_data
purchase_count = purchase_data.groupby('Gender')['Price'].count()

# series called average_purchase_price
average_purchase_price = purchase_data.groupby('Gender')['Price'].mean()

# series called total_purchase_value
# total_purchase_value = purchase_data.groupby(['Gender'])
#total_purchase_value = pd.DataFrame(purchase_data.groupby(['Gender']).sum()['Price'])
total_purchase_value = purchase_data.groupby('Gender')['Price'].sum()
total_purchase_value

# series called avgerage_total_person
average_total_person = total_purchase_value / player_demographics_count['Total Count']

# building the dataframe
purchasing_analysis = pd.DataFrame({'Purchase Count': purchase_count, 
                                    'Average Purchase Price': average_purchase_price,
                                    'Total Purchase Value': total_purchase_value, 
                                    'Avg Total Purchase per Person': average_total_person
                                   })
purchasing_analysis

# problem 1: can't get average_total_person to work
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# create new dataframe age_demos_df from original dataframe purchase_data
age_demos_df = purchase_data
# age_demos_df.max()

# creating bins for the age groups
bins = [0, 9.99, 14, 19, 24, 29, 34, 39, age_demos_df['Age'].max()]
# print(bins)

# creating group names to associate with the bins
age_bracket = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

# making a new column in age_demos_df to hold the age bins
age_demos_df['Age Bracket'] = pd.cut(age_demos_df['Age'], bins, labels=age_bracket, include_lowest=True)
age_demos_df = age_demos_df[['Age Bracket', 'SN']]
age_demos_df

# making a new dataframe with the duplicate screen names removed 
# age_demos_df = age_demos_df.set_index('SN')
age_demos_dropped_df = age_demos_df.drop_duplicates()
# age_demos_dropped_df = age_demos_df['Age Bracket']
age_demos_dropped_df['Age Bracket'].value_counts()
age_demos_dropped_df = age_demos_dropped_df['Age Bracket'].value_counts()
age_demos_dropped_df.sort_values(axis=0, ascending=False)

percent_players = age_demos_dropped_df / total_players * 100
percent_players

final_age_demo_df = pd.DataFrame({'Total Count': age_demos_dropped_df, 'Percentage of Players': percent_players})
final_age_demo_df['Percentage of Players'] = final_age_demo_df['Percentage of Players'].map("{:.2f}%".format)
final_age_demo_df.sort_index()
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# create new dataframe age_demos_df from original dataframe purchase_data
age_purchase_df = purchase_data[['Age','Purchase ID', 'Price', 'SN']]
# age_demos_df.max()

# creating bins for the age groups
bins = [0, 9.99, 14, 19, 24, 29, 34, 39, age_purchase_df['Age'].max()]
# print(bins)

# creating group names to associate with the bins
age_bracket = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

# making a new column in age_demos_df to hold the age bins
age_purchase_df['Age Ranges'] = pd.cut(age_purchase_df['Age'], bins, labels=age_bracket, include_lowest=True)
age_purchase_df

# making a new series for 'Purchase Count'
age_purchase_count = age_purchase_df.groupby(['Age Ranges']).count()["Price"]

age_average_purchase_price = age_purchase_df.groupby(['Age Ranges']).mean()["Price"]

age_purchase_total = age_purchase_df.groupby(['Age Ranges']).sum()["Price"]

player_count = purchase_data['SN'].nunique()
age_average_total_person = age_purchase_total / age_demos_dropped_df


final_purchase_analysis_df = pd.DataFrame({"Purchase Count": age_purchase_count, 
                                        "Average Purchase Price": age_average_purchase_price, 
                                        "Total Purchase Value": age_purchase_total, 
                                        "Avg Total Purchase per Person": age_average_total_person})
final_purchase_analysis_df
```
